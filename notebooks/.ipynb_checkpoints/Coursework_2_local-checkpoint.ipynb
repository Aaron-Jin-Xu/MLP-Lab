```python
import numpy as np
import matplotlib.pyplot as plt
import logging
%matplotlib inline
plt.style.use('ggplot')
%load_ext autoreload
%autoreload 2

logger = logging.getLogger()
logger.setLevel("INFO") 
logger.handlers = [logging.StreamHandler()] 
seed = 20161121
rng = np.random.RandomState(seed)
```

```python
from mlp.models import MultipleLayerModel
from mlp.layers import AffineLayer, ConvolutionalLayer, MaxPoolingLayer, ReshapeLayer
from mlp.layers import SigmoidLayer, TanhLayer, ReluLayer, DropoutLayer
from mlp.data_providers import MNISTDataProvider, MNISTCNNAutoencoderDataProvider, MNISTCNNDataProvider
from mlp.data_providers import MNISTCNNDenoisingAutoencoderDataProvider
from mlp.errors import CNNSumOfSquaredDiffsError, CrossEntropySoftmaxError
from mlp.learning_rules import GradientDescentLearningRule, MomentumLearningRule
from mlp.schedulers import ConstantLearningRateScheduler, ExponentialLearningRateScheduler
from mlp.penalties import L1Penalty, L2Penalty
from mlp.initialisers import GlorotUniformInit, ConstantInit, UniformInit
from mlp.optimisers import Optimiser
from mlp.settings import Setting
from mlp.experiments import Experiment
```

```python
train_set = MNISTCNNDataProvider(which_set='train', batch_size=100, max_num_batches=-1, rng=rng)
valid_set = MNISTCNNDataProvider(which_set='valid', batch_size=100, max_num_batches=-1, rng=rng)
```

```python
kernels_init = UniformInit(-0.01,0.01,rng=rng)
weights_init = GlorotUniformInit(rng=rng)
biases_init = ConstantInit(0.)

learning_rule = MomentumLearningRule(learning_rate=0.02, mom_coeff=0.9)
error = CrossEntropySoftmaxError()
ae_error = CNNSumOfSquaredDiffsError()
```

```python
# Model Architecture
import time


layers_1 = [
    ConvolutionalLayer(num_input_channels=1, num_output_channels=4, 
                       input_dim_1=28, input_dim_2=28, 
                       kernel_dim_1=5, kernel_dim_2=5),
    TanhLayer(),
    MaxPoolingLayer(),
    ReshapeLayer(),
    AffineLayer(input_dim=12*12*4, output_dim=100, 
                weights_initialiser=weights_init, biases_initialiser=biases_init),
    TanhLayer(),
    AffineLayer(input_dim=100, output_dim=10, 
                weights_initialiser=weights_init, biases_initialiser=biases_init)
]

model_1 = MultipleLayerModel(layers_1)
setting_1 = Setting(model_1, error, learning_rule, train_set, valid_set)

# Model Architecture

layers_2 = [
    ConvolutionalLayer(num_input_channels=1, num_output_channels=4, 
                       input_dim_1=28, input_dim_2=28, 
                       kernel_dim_1=5, kernel_dim_2=5),
    ReluLayer(),
    MaxPoolingLayer(),
    ReshapeLayer(),
    AffineLayer(input_dim=12*12*4, output_dim=100, 
                weights_initialiser=weights_init, biases_initialiser=biases_init),
    ReluLayer(),
    AffineLayer(input_dim=100, output_dim=10, 
                weights_initialiser=weights_init, biases_initialiser=biases_init)
]

model_2 = MultipleLayerModel(layers_2)
setting_2 = Setting(model_2, error, learning_rule, train_set, valid_set)


# Model Architecture

layers_3 = [
    ConvolutionalLayer(num_input_channels=1, num_output_channels=4, 
                       input_dim_1=28, input_dim_2=28, 
                       kernel_dim_1=5, kernel_dim_2=5),
    SigmoidLayer(),
    MaxPoolingLayer(),
    ReshapeLayer(),
    AffineLayer(input_dim=12*12*4, output_dim=100, 
                weights_initialiser=weights_init, biases_initialiser=biases_init),
    SigmoidLayer(),
    AffineLayer(input_dim=100, output_dim=10, 
                weights_initialiser=weights_init, biases_initialiser=biases_init)
]

model_3 = MultipleLayerModel(layers_3)
setting_3 = Setting(model_3, error, learning_rule, train_set, valid_set)


layers_4 = [
    ConvolutionalLayer(num_input_channels=1, num_output_channels=4, 
                       input_dim_1=28, input_dim_2=28, 
                       kernel_dim_1=5, kernel_dim_2=5),
    MaxPoolingLayer(),
    ReshapeLayer(),
    AffineLayer(input_dim=12*12*4, output_dim=100, 
                weights_initialiser=weights_init, biases_initialiser=biases_init),
    AffineLayer(input_dim=100, output_dim=10, 
                weights_initialiser=weights_init, biases_initialiser=biases_init)
]

model_4 = MultipleLayerModel(layers_4)
setting_4 = Setting(model_4, error, learning_rule, train_set, valid_set)

settings = [setting_1, setting_2, setting_3, setting_4]

remark = "Compare Non-linear Transformation Function"
experiment = Experiment(settings=settings, num_epoch=20, stats_interval=1, remark=remark, 
                        legends=['Tanh', 'Relu', 'Sigmoid', 'None'])
experiment.do_experiment()
```

```python
_,_,_,_, tanh = experiment.load_results("23-18_12_18", 1)
_,_,_,_, relu = experiment.load_results("23-18_12_18", 2)
_,_,_,_, sigm = experiment.load_results("23-18_12_18", 3)
_,_,_,_, none = experiment.load_results("23-18_12_18", 4)

layer = 6


x = []
y = []
e = []
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
for i,epoch in enumerate(tanh):
    a = epoch[layer]
    x.append(i)
    y.append(np.mean(a))
    e.append(np.std(a))
ax.errorbar(x, y, e, linestyle='-', marker='^')

x = []
y = []
e = []
for i,epoch in enumerate(relu):
    a = epoch[layer]
    x.append(i)
    y.append(np.mean(a))
    e.append(np.std(a))
ax.errorbar(x, y, e, linestyle='-', marker='^')
x = []
y = []
e = []
for i,epoch in enumerate(sigm):
    a = epoch[layer]
    x.append(i)
    y.append(np.mean(a))
    e.append(np.std(a))
ax.errorbar(x, y, e, linestyle='-', marker='^')

x = []
y = []
e = []
for i,epoch in enumerate(none):
    a = epoch[layer-2]
    x.append(i)
    y.append(np.mean(a))
    e.append(np.std(a))
ax.errorbar(x, y, e, linestyle='-', marker='^')


ax.legend(["Tanh","Relu","Sigmoid","None"])

fig.savefig("activation-2.pdf")

plt.show()
```
